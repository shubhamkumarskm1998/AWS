# 🚀 High Availability, Scaling, Load Balancing & Auto Scaling (AWS EC2)

---

# 🧱 1. Vertical Scaling (Scale Up / Down)

## 💡 What is it?

👉 Increasing the **size of a single EC2 instance**

---

## 📌 Example

- Small → t2.nano (0.5 GB RAM)
- Large → u-12tb1.metal (12.3 TB RAM)

---

## 🧠 Real-Life Example

👉 You have a **database server**

- Initially: small instance
- Traffic increases → upgrade to bigger machine

---

## ⚠️ Limitations

- Has a **maximum limit**
- Requires **downtime (restart)**
- Single point of failure ❌

---

# 🌐 2. Horizontal Scaling (Scale Out / In)

## 💡 What is it?

👉 Increasing **number of EC2 instances**

---

## 🧠 Real-Life Example

👉 Website like an e-commerce app:

- 1 server → slow
- Add 5 servers → faster

---

## 📌 Components

- Auto Scaling Group (ASG)
- Load Balancer

---

## ✅ Benefits

- No single point of failure
- High availability
- Better performance

---

# 🏥 3. High Availability (HA)

## 💡 What is it?

👉 Application runs across **multiple AZs**

---

## 🧠 Real-Life Example

👉 If one data center fails:

- Other AZ still serves traffic

---

## 📌 Setup

- ASG across multiple AZs
- Load Balancer distributes traffic

---

## 🎯 Result

👉 Zero downtime system

---

# ⚖️ 4. Scalability vs Elasticity vs Agility

---

## 📈 Scalability

👉 Handle more load

- Scale UP (bigger server)
- Scale OUT (more servers)

---

## ⚡ Elasticity

👉 Auto adjust based on demand

---

## 🧠 Real-Life Example

👉 Food delivery app:

- Lunch time → scale up
- Night → scale down

---

## 🚀 Agility

👉 Launch resources quickly

---

## 🧠 Example

👉 Create 10 servers in minutes (instead of weeks)

---

# 🔄 5. What is Load Balancing?

## 💡 Definition

👉 Distributes traffic across multiple EC2 instances

---

## 🧠 Real-Life Example

👉 Restaurant:

- One waiter = overload
- Multiple waiters = balanced service

---

👉 Load balancer = manager assigning customers

---

# 🎯 6. Why use Load Balancer?

---

## ✅ Benefits

- Distributes traffic
- Single entry point (DNS)
- Health checks
- Fault tolerance
- SSL termination (HTTPS)
- Multi-AZ support

---

# ☁️ 7. Elastic Load Balancer (ELB)

## 💡 What is it?

👉 AWS managed load balancer

---

## ✅ Benefits

- AWS manages infra
- Auto scaling
- High availability

---

# 🔥 8. Types of Load Balancers

---

## 🌍 1. Application Load Balancer (ALB)

👉 Layer 7 (HTTP/HTTPS)

### 📌 Features

- URL-based routing
- Host-based routing

---

## 🧠 Example

👉 Route:

- `/api` → backend
- `/images` → CDN

---

## ⚡ 2. Network Load Balancer (NLB)

👉 Layer 4 (TCP/UDP)

### 📌 Features

- Ultra high performance
- Static IP

---

## 🧠 Example

👉 Gaming server or real-time apps

---

## 🔐 3. Gateway Load Balancer (GWLB)

👉 Layer 3

### 📌 Use Case

- Security appliances
- Firewalls

---

## ⚠️ 4. Classic Load Balancer

👉 Old (deprecated)

---

# ⚙️ 9. Auto Scaling Group (ASG)

---

## 💡 What is it?

👉 Automatically manages EC2 instances

---

## 🎯 Goals

- Add/remove instances
- Maintain min/max count
- Replace unhealthy instances

---

## 🧠 Real-Life Example

👉 E-commerce app:

- Normal → 2 servers
- Sale → 20 servers
- After sale → back to 2

---

# 🔁 10. ASG + Load Balancer Flow

---

👉 User → Load Balancer → Multiple EC2 → Auto Scaling adjusts

---

## 📌 Key Feature

- Auto-register instances to LB
- Remove unhealthy instances

---

# 💰 11. Cost Benefit

👉 Pay only for usage

# 📊 12. Scaling Strategies

---

## 🧍 Manual Scaling

👉 Manually change instance count

---

## 🔄 Dynamic Scaling

---

### 1. Simple / Step Scaling

👉 Based on CloudWatch alarms

---

## 🧠 Example

- CPU > 70% → add 2 instances
- CPU < 30% → remove 1

---

### 2. Target Tracking

👉 Maintain specific metric

---

## 🧠 Example

👉 Keep CPU ~40%

---

### 3. Scheduled Scaling

👉 Based on time

---

## 🧠 Example

👉 Every Friday 5 PM → increase servers

---

## 🤖 4. Predictive Scaling

👉 Uses ML to predict traffic

---

## 🧠 Real-Life Example

👉 Netflix-like app:

- Evening traffic always high
- AWS adds servers before spike

---

## 🎯 Best Use Case

👉 Predictable workloads

---

---

- ASG = auto manage instances
- Elasticity = auto scale
- Predictive scaling = ML based
