# 🔐 AWS IAM

---

# 🧠 What is IAM?

IAM = **Identity and Access Management**

👉 Used to:
- Manage users
- Control access
- Secure AWS resources

---

# 👤 1. Users

---

## 💡 What is a User?

👉 A **user = a person or application** using AWS

---

## 📌 Key Points

- Has:
  - Username
  - Password (for console login)
  - Access keys (for CLI/SDK)

---

## 🧠 Example

👨‍💻 Company has 2 employees:
- Rahul → IAM User
- Priya → IAM User

Each gets:
- Login to AWS Console
- Permissions based on role

---

## 🎯 Real Scenario

- Developer → access EC2
- Analyst → access S3 only

---

# 👥 2. Groups

---

## 💡 What is a Group?

👉 A **collection of users**

---

## 📌 Key Points

- Contains only users
- Permissions assigned to group
- All users inherit same permissions

---

## 🧠 Example

Group: `Developers`

Users:
- Rahul
- Priya

Permission:
- Full EC2 access

👉 Both automatically get access

---

## 🎯 Benefit

👉 No need to assign permissions individually

---

# 📜 3. Policies

---

## 💡 What is a Policy?

👉 A **JSON document** that defines permissions

---

## 📌 Structure

Policy defines:
- What action?
- On which resource?
- Allowed or denied?

---

## 🧠 Example Policy

```json
{
  "Effect": "Allow",
  "Action": "s3:ListBucket",
  "Resource": "*"
}

## 🎯 Meaning
- Allow listing S3 buckets
- On all resources

---

## 🔥 Real Example

👉 Give read-only access to S3:

- Action → GetObject
- Effect → Allow

---

# 🎭 4. Roles

## 💡 What is a Role?
👉 A temporary permission assigned to AWS services or users

---

## 📌 Key Points
- No username/password
- Used by:
  - EC2
  - Lambda
  - Other AWS services

---

## 🧠 Example

👉 EC2 instance needs access to S3

Instead of storing keys:

- Attach IAM Role to EC2

---

## 🎯 Benefit
👉 More secure (no hardcoded credentials)

---

# 🔐 5. Security (MFA + Password Policy)

---

## 🔑 Password Policy

👉 Rules for passwords:

- Minimum length
- Special characters
- Expiry

---

## 🔒 MFA (Multi-Factor Authentication)

👉 Login requires:

- Password
- OTP (mobile app)

---

## 🧠 Example

Login requires:

- Password + Google Authenticator code

---

## 🎯 Benefit
👉 Extra security layer

---

# 💻 6. AWS CLI

---

## 💡 What is CLI?
👉 Command Line Interface to manage AWS

---

## 🧠 Example

```bash
aws s3 ls
```

👉 Lists all S3 buckets

---

## 🎯 Use Case
- Automation
- DevOps scripts

---

# 🧑‍💻 7. AWS SDK

---

## 💡 What is SDK?
👉 Use programming language to interact with AWS

---

## 🧠 Example (Python)

```python
import boto3

s3 = boto3.client('s3')
response = s3.list_buckets()
print(response)
```

---

## 🎯 Use Case
- Build applications
- Backend integration

---

# 🔑 8. Access Keys

---

## 💡 What are Access Keys?
👉 Used for CLI/SDK access

---

## 📌 Components
- Access Key ID
- Secret Access Key

---

## 🧠 Example

Used in:

```bash
aws configure
```

---

## ⚠️ Important

👉 Never share keys publicly
👉 Store securely

---

# 📊 9. Audit (Monitoring IAM)

---

## 📌 Tools

---

### 📄 IAM Credential Report

👉 Shows:

- Who has password?
- MFA enabled or not?

---

### 📊 IAM Access Advisor

👉 Shows:

- Which services user actually used

---

## 🧠 Example

User has:

- EC2 + S3 access

But uses only S3

👉 Remove EC2 permission (security best practice)
