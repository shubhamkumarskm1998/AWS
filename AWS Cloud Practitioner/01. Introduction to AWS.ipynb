# 1. What is Cloud Computing?

Cloud Computing means:

👉 Instead of storing data or running applications on your own computer or servers,
👉 You use servers available on the **internet (cloud)**.

💡 Simple meaning:
- Your data + apps are stored somewhere else (data centers)
- You access them using the internet

📌 Example:
- Google Drive
- Netflix
- AWS / Azure

---

# 2. How Websites Work (Basic Flow)

A website works using 3 main things:

### 🖥️ Client
- Your laptop / mobile
- You open browser and request a website

### 🌐 Network (Internet)
- Connects client to server
- Data travels through internet

### 🖥️ Server
- A powerful computer
- Stores website data and logic

---

## 🔁 Flow (Step-by-step)

1. You type: `www.google.com`
2. Request goes through internet
3. Server receives request
4. Server sends response (HTML, data)
5. Browser shows website

---

## 📌 Important Concept: IP Address

- Every device has an **IP address**
- Like a phone number

👉 Clients → Have IP
👉 Servers → Have IP

# 💰 1. Pricing of the Cloud (AWS Pricing Model)

AWS follows **pay-as-you-go pricing**
👉 You only pay for what you use

---

## 🔑 3 Main Pricing Fundamentals

---

## ⚙️ 1. Compute Pricing

👉 You pay for **compute time**

### What it means:
- If a server runs for 1 hour → you pay for 1 hour
- If stopped → no cost

### Example:
- EC2 instance running → billed
- Lambda function → billed per execution time

---

## 💾 2. Storage Pricing

👉 You pay for **data stored in cloud**

### What it means:
- More storage → more cost
- Less storage → less cost

### Example:
- S3 bucket storing 100GB → charged for 100GB

---

## 🌐 3. Data Transfer Pricing

👉 **Data IN (upload)** → FREE
👉 **Data OUT (download)** → PAID

### Example:
- Upload file to AWS → free
- Download file → charged

---

## 🎯 Key Benefit

👉 Solves traditional IT cost problems:
- No hardware cost
- No maintenance cost
- No over-provisioning

# 🔐 AWS Shared Responsibility Model

---

# 🧠 First Understand This (MOST IMPORTANT)

👉 Security in AWS is **shared between YOU and AWS**

---

## 🔥 Simple Rule

👉 AWS = **Security OF the Cloud**
👉 You = **Security IN the Cloud**

---

## 🧠 Meaning

- AWS protects **infrastructure**
- You protect **your data & applications**

---

# 🏗️ 1. AWS Responsibility (Security OF the Cloud)

👉 AWS handles everything **below the line (infrastructure)**

---

## 🧱 What AWS Manages

---

## 🖥️ Hardware / Infrastructure

- Data centers
- Physical servers
- Storage devices
- Networking cables

---

## 🌍 Global Infrastructure

- Regions
- Availability Zones
- Edge Locations

---

## ⚙️ Core Services

- Compute (EC2 hardware)
- Storage (S3 hardware)
- Databases (physical infra)
- Networking

---

## 🔐 Security at AWS Level

- Physical security (guards, CCTV)
- Power & cooling
- Hardware maintenance
- Network protection

---

## 🧠 Simple Meaning

👉 AWS ensures:
> “Cloud is safe and running properly”

---

# 👨‍💻 2. Customer Responsibility (Security IN the Cloud)

👉 YOU manage everything you put inside AWS

---

## 📦 What YOU Manage

---

## 🗄️ 1. Customer Data

- Files
- Databases
- Backups

👉 You must:
- Protect it
- Encrypt it

---

## 🔐 2. Identity & Access Management (IAM)

- Who can access what
- Users, roles, permissions

---

## 🧩 3. Applications

- Your code
- APIs
- Business logic

---

## 🖥️ 4. Platform & Configuration

- App settings
- Services configuration

---

## ⚙️ 5. Operating System (in IaaS)

- OS updates
- Patches
- Security configs

---

## 🔥 6. Network & Firewall Configuration

- Security Groups
- VPC rules

---

## 🔒 7. Encryption Responsibilities

---

### Client-side Encryption
- Encrypt data before sending to AWS

---

### Server-side Encryption
- Enable encryption in S3, EBS, etc.

---

### Network Encryption
- Use HTTPS / SSL

---

## 🧠 Simple Meaning

👉 You ensure:
> “Your app and data are सुरक्षित (secure)”
