# 🚀 EC2 Detailed Notes

---

# 🧠 What is EC2?

EC2 (Elastic Compute Cloud) = Virtual server in AWS

- Launch servers
- Run applications
- Scale anytime

---

# 🧱 EC2 Instance Components

EC2 Instance =

- AMI (OS)
- Instance Size (CPU + RAM)
- Storage
- Security Groups
- EC2 User Data

---

# 🖥️ AMI (Amazon Machine Image)

- Pre-configured OS
- Examples:
  - Ubuntu
  - Amazon Linux
  - Windows

Important:
- Region specific
- Can create custom AMI

---

# ⚙️ Instance Size

Defines:
- CPU
- RAM

Examples:
- t2.micro → small apps
- m5.large → medium apps
- c5.xlarge → CPU heavy

Important:
- Can resize later
- Bigger = more cost

---

# 💾 Storage

Types:

1. EBS (Elastic Block Store)
- Persistent
- Survives restart

2. Instance Store
- Temporary
- Data lost on stop

Important:
- EBS supports snapshot
- Can attach/detach

---

# 🔥 Security Groups

Firewall for EC2

Controls:
- Inbound traffic
- Outbound traffic

Example:
- SSH → port 22
- HTTP → port 80

IMPORTANT:

- All inbound → BLOCKED by default
- All outbound → ALLOWED by default
- Stateful
- No DENY rules (only ALLOW)
- Can attach to multiple EC2
- EC2 can have multiple SG

Best Practice:
- Least privilege (open only required ports)

---

# 📜 EC2 User Data

- Script runs at first startup

Example:

#!/bin/bash
apt update
apt install -y nginx

Important:
- Runs once (default)
- Used for automation

---

# 🔐 SSH

- Connect to EC2 terminal
- Port 22
- Uses .pem key

Example:

ssh -i mykey.pem ubuntu@<public-ip>

Important:
- No password login (default)
- Keep key safe

---

# 🎭 EC2 Instance Role

- IAM Role attached to EC2

Example:
- EC2 accessing S3

Important:
- No hardcoded keys
- Temporary credentials
- More secure

---

# 💰 Purchasing Options

1. On-Demand
- Pay per usage
- Flexible
- Expensive

2. Spot
- Very cheap
- Can terminate anytime

3. Reserved
- 1–3 year commitment
- Cheaper

Types:
- Standard
- Convertible

4. Dedicated Host
- Full physical server
- For compliance

5. Dedicated Instance
- Dedicated hardware

---

# 🎯 FINAL IMPORTANT POINTS

- EC2 = Virtual server
- AMI = OS
- EBS = persistent storage
- Security Group = firewall

MOST IMPORTANT:

- SG blocks all inbound by default
- SG allows all outbound by default
- SG is stateful
- SG supports least privilege

---

# 🚀 ONE LINE

EC2 = Customizable virtual server in AWS


# 💾 EC2 Instance Storage – Detailed Notes + Important Points

---

# 🧠 Big Picture

AWS provides **multiple storage options** for EC2:

- EBS → Persistent disk (most common)
- Instance Store → Temporary disk (very fast)
- EFS → Shared file system
- FSx → Specialized file systems
- AMI → Pre-built machine image
- EBS Snapshots → Backup mechanism

---

# 💾 1. EBS (Elastic Block Store)

---

## 💡 What is EBS?

👉 Network-attached disk for EC2

---

## 📌 Key Features

- Attached to **one EC2 at a time**
- Lives in a **single Availability Zone (AZ)**
- Persistent storage (data survives restart)

---

## 🧠 Example

👉 EC2 running a web app:

- OS stored on EBS
- Database stored on EBS

---

## ⚠️ IMPORTANT POINTS

- AZ-specific (cannot directly move across AZ)
- Detach and attach supported
- Data persists after stop
- Supports snapshots (backup)

---

# 📸 2. EBS Snapshots (VERY IMPORTANT)

---

## 💡 What is EBS Snapshot?

👉 Backup of an EBS volume stored in S3

---

## 📌 Key Features

- Incremental backup (only changed data saved)
- Stored in S3 (managed by AWS)
- Can recreate volume anytime

---

## 🧠 Example

👉 Take snapshot of 20GB volume:

- Only changed data stored → saves cost

---

## 🔄 Use Cases

- Backup data
- Disaster recovery
- Copy data across AZ/regions
- Create AMI

---

## ⚠️ IMPORTANT POINTS (EXAM)

---

### 🔁 Incremental Backup

👉 Only changes since last snapshot are stored

---

### 🌍 Cross-AZ / Cross-Region

👉 Snapshots allow:

- Restore in another AZ
- Copy to another region

---

### 🚀 Restore

👉 Create new EBS volume from snapshot

---

### 🔒 Safety

👉 Even if EC2 is deleted → snapshot safe

---

### 💰 Cost

👉 Pay only for stored data

---

### 🎯 Best Practice

👉 Take **regular snapshots** for backup

---

# 🧱 3. AMI (Amazon Machine Image)

---

## 💡 What is AMI?

👉 Template used to launch EC2 instances

---

## 📌 Includes

- OS
- Software
- Configurations

---

## 🧠 Example

👉 Create AMI:

- Install app
- Save image
- Launch multiple servers

---

## ⚠️ IMPORTANT

- AMI uses **EBS snapshots internally**
- Region-specific

---

# ⚙️ 4. EC2 Image Builder

---

## 💡 What is it?

👉 Automatically builds and updates AMIs

---

## 🧠 Example

- Weekly updated AMI with patches

---

## ⚠️ Important

- Ensures consistency
- Improves security

---

# ⚡ 5. EC2 Instance Store

---

## 💡 What is it?

👉 Local disk attached to EC2

---

## 📌 Features

- Very fast
- Temporary

---

## ⚠️ CRITICAL

- Data LOST on stop/terminate

---

## 🧠 Use Case

- Cache
- Temporary processing

---

# 🌐 6. EFS (Elastic File System)

---

## 💡 What is EFS?

👉 Shared file system

---

## 📌 Features

- Multi-instance access
- Multi-AZ
- Auto scaling

---

## 🧠 Example

- Shared storage for web servers

---

## ⚠️ Important

- Expensive compared to EBS
- Good for shared workloads

---

# 💸 7. EFS-IA

---

## 💡 What is it?

👉 Cheaper storage for infrequent access

---

## ⚠️ Trade-off

- Lower cost
- Slightly slower

---

# 🪟 8. FSx for Windows

---

## 💡 What is it?

👉 Windows file system

---

## 📌 Features

- SMB protocol
- Active Directory integration

---

## 🧠 Use Case

- Enterprise Windows apps

---

# 🧪 9. FSx for Lustre

---

## 💡 What is it?

👉 High-performance file system

---

## 🧠 Use Case

- Machine Learning
- HPC workloads

---

# 🔥 FINAL COMPARISON

---

| Storage | Type | Persistence | Multi-Instance | Use Case |
|--------|------|------------|---------------|---------|
| EBS | Block | ✅ Yes | ❌ No | General |
| Snapshot | Backup | ✅ Yes | - | Backup/restore |
| Instance Store | Block | ❌ No | ❌ No | Temp |
| EFS | File | ✅ Yes | ✅ Yes | Shared |
| FSx | File | ✅ Yes | ✅ Yes | Specialized |
| AMI | Template | ✅ Yes | - | Launch |

---

---

- EBS → AZ specific
- Snapshot → stored in S3
- Snapshot → incremental
- Instance Store → data lost
- EFS → multi-AZ
- AMI → built from snapshots
