# 🪣 3. S3 Buckets

---

## 💡 What is a Bucket?

👉 A **container to store objects (files)**

---

## 📌 Important Rules

- Must be **globally unique**
- Created at **region level**
- S3 looks global but bucket is regional

---

## 🧠 Naming Rules

- No uppercase
- No underscore
- Length: 3–63 characters
- Must not be an IP
- Start with lowercase or number
- Cannot start with `xn--`
- Cannot end with `-s3alias`

---

## 🧠 Real-Life Example

👉 Bucket = Folder

```
my-photo-bucket
```

---

# 📦 4. S3 Objects

---

## 💡 What is an Object?

👉 A **file stored in S3**

---

## 📌 Key Concept

👉 Every object has a **KEY (full path)**

```
s3://my-bucket/file.txt
s3://my-bucket/folder1/folder2/file.txt
```

---

## ⚠️ Important

👉 There are **NO real folders**

- Just long names with `/`

---

## 🧠 Example

👉 UI shows folders but actually:

```
folder1/file.txt → just a string
```

---

# 📦 5. Object Details

---

## 📌 Object Size

- Max size = **5 TB**
- >5GB → use **Multipart Upload**

---

## 📌 Metadata

👉 Key-value pairs

```
Content-Type: image/png
```

---

## 📌 Tags

👉 Used for:

- Security
- Lifecycle management

---

## 📌 Version ID

👉 Exists if versioning is enabled

---

# 🔐 6. S3 Security

---

## 🔹 1. User-Based (IAM)

👉 Control access using IAM policies

---

## 🔹 2. Resource-Based

### Bucket Policies

- Applied to entire bucket
- Supports cross-account access

---

### ACL (Access Control List)

- Object-level permissions
- Less used (can disable)

---

## ⚠️ Important Rule

👉 Access is allowed if:

- IAM allows ✅ OR
- Bucket policy allows ✅
- AND no explicit DENY ❌

---

## 🔒 Encryption

👉 Data can be encrypted:

- At rest
- In transit

---

# 📜 7. S3 Bucket Policies

---

## 💡 JSON Based Policies

---

## 📌 Components

- Resource → bucket/object
- Effect → Allow / Deny
- Action → API actions
- Principal → who gets access

---

## 🧠 Example

👉 Public read access:

```json
{
  "Effect": "Allow",
  "Principal": "*",
  "Action": ["s3:GetObject"],
  "Resource": ["arn:aws:s3:::my-bucket/*"]
}
```

---

## 🎯 Use Cases

- Public website hosting
- Force encryption
- Cross-account access

---

# 🔄 9. Versioning

---

## 💡 What is it?

👉 Store multiple versions of same file

---

## 📌 Key Points

- Enabled at bucket level
- Same key → new version

---

## 🎯 Benefits

- Recover deleted files
- Rollback changes

---

## ⚠️ Important Notes

- Old files get version "null"
- Disabling versioning does NOT delete versions

---

## 🧠 Example

```
file.txt → v1
file.txt → v2
file.txt → v3
```

---

# 💾 10. Storage Classes

---

## 📦 Types

### 🟢 S3 Standard
- Frequent access
- High availability

### 🟡 S3 IA (Infrequent Access)
- Less frequent access
- Lower cost

### 🟠 One Zone IA
- Stored in 1 AZ
- Cheaper but less resilient

### 🧊 Glacier (Archive)
- Very cheap
- Slow retrieval

#### Types:
- Glacier Instant
- Glacier Flexible
- Glacier Deep Archive

### 🤖 Intelligent Tiering
- Auto moves data based on usage

---

## 🔁 Lifecycle

👉 Move data automatically between classes

---

# 💪 11. Durability vs Availability

---

## 🛡️ Durability

👉 Data safety

- 99.999999999% (11 9’s)
- Multi-AZ replication

---

## ⚡ Availability

👉 How often data is accessible

📌 Example:
- 99.99% availability
- ~53 minutes downtime/year

---

# 🤖 12. Intelligent Tiering

---

## 💡 What is it?

👉 Automatically moves data between tiers

---

## 📌 Features

- Small monitoring fee
- No retrieval charges

---

## 📊 Tiers

- Frequent access
- Infrequent access (30 days)
- Archive instant (90 days)
- Archive (90–700 days)
- Deep archive (180+ days)

---

## 🧠 Real-Life Example

👉 User uploads files:

- Frequently used → fast tier
- Old files → moved to cheap storage

---

---

- Object = file
- Key = full path
- No real folders
- Max size = 5TB
- Multipart upload >5GB
- IAM + Bucket Policy control access

# 🚀 AWS S3 Advanced Topics — Complete Notes

# 🔐 1. S3 Encryption

---

## 🔹 Server-Side Encryption (Default)

👉 AWS encrypts data AFTER upload

### Types:
- SSE-S3 → AWS managed keys
- SSE-KMS → KMS keys (more control)
- SSE-C → Customer provides key

### ✅ Pros:
- Easy to use
- No key management (SSE-S3)

---

## 🔹 Client-Side Encryption

👉 Data encrypted BEFORE upload

### ✅ Pros:
- Full control
- AWS cannot see data

### ❌ Cons:
- You manage keys
- More complexity

---

## 🧠 Real-Life Example

- SSE → Upload photo → AWS encrypts
- Client-side → Encrypt file locally → upload

---

# ⚡ 2. S3 Express One Zone

---

## 💡 What is it?

👉 High-performance S3 storage in **single AZ**

---

## 📌 Key Features

- Ultra-low latency (ms)
- Handles 100,000+ requests/sec
- Up to 10x faster than S3 Standard
- Lower cost

---

## ⚠️ Trade-off

- Stored in ONE AZ → less resilient

---

## 🧠 Use Cases

- AI/ML training
- Financial analytics
- Real-time apps

# 📦 5. AWS Snowball

---

## 💡 What is it?

👉 Physical device for **data transfer**

---

## 📌 Why?

- Internet too slow
- Large data (TB/PB)

---

## 📊 Capacity

- Up to petabytes
- Compute + Storage versions

---

## 🧠 Example

👉 100 TB migration:

- Internet → weeks
- Snowball → days
