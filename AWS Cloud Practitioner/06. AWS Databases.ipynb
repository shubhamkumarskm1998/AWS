# 🚀 AWS Databases — Complete Notes

---

# 🧠 Databases & Shared Responsibility on AWS

---

## 💡 Overview

AWS provides managed database services to simplify operations and scaling.

---

## ✅ Benefits of Managed Databases

- ⚡ Quick provisioning
- 📈 High availability
- 🔄 Vertical & horizontal scaling
- 💾 Automated backups & restore
- 🔧 Automated operations & upgrades
- 🖥️ OS patching handled by AWS
- 📊 Monitoring & alerting

---

## ⚠️ Important Note

👉 If you run DB on EC2:
- You manage backups
- You manage patching
- You manage HA & scaling

👉 If you use managed DB (RDS, DynamoDB):
- AWS handles most of it

---

# 🗄️ Amazon RDS Overview

---

## 💡 What is RDS?

- Relational Database Service
- Managed SQL database

---

## 📌 Supported Engines

- PostgreSQL
- MySQL
- MariaDB
- Oracle
- SQL Server
- IBM DB2
- Aurora

---

# ⚡ RDS vs EC2 (DB on EC2)

---

## ✅ RDS Advantages

- Automated provisioning
- OS patching
- Point-in-time restore
- Monitoring dashboards
- Read replicas
- Multi-AZ for DR
- Maintenance windows
- Auto scaling
- Storage backed by EBS

---

## ❌ Limitation

- No SSH access to DB instance

---

# 🏗️ RDS Architecture

---

## 📌 Typical Setup

- Load Balancer
- EC2 instances (App layer)
- RDS database (backend)

---

# 🚀 Amazon Aurora

---

## 💡 What is Aurora?

- AWS proprietary DB
- Compatible with MySQL & PostgreSQL

---

## ⚡ Key Features

- 5x faster than MySQL
- 3x faster than PostgreSQL
- Auto-scaling storage (up to 256 TB)
- Highly optimized for cloud

---

## 💰 Cost

- ~20% more expensive than RDS
- But more efficient

---

# ⚡ Aurora Serverless

---

## 💡 What is it?

- Auto-scaling DB based on usage

---

## ✅ Features

- No capacity planning
- Pay per second
- Minimal management

---

## 🧠 Use Cases

- Intermittent workloads
- Unpredictable traffic

---

# 🔄 RDS Deployments

---

## 📖 Read Replicas

- Scale read traffic
- Up to 15 replicas
- Writes only to main DB

---

## 🌍 Multi-AZ

- Used for high availability
- Automatic failover
- One standby replica

---

# 🌎 Multi-Region (Global Read Replicas)

---

## 💡 Purpose

- Disaster recovery
- Global low latency

---

## ⚠️ Trade-off

- Additional cost
- Replication delay

---

# ⚡ Amazon ElastiCache

---

## 💡 What is it?

- Managed Redis / Memcached

---

## 🚀 Benefits

- In-memory caching
- Ultra-low latency
- Reduces DB load

---

## 🧠 Use Case

- Read-heavy applications
- Session storage
- Caching API responses

---

# 🏗️ ElastiCache Architecture

---

## Flow

- User → Load Balancer → EC2
- EC2 → Cache (fast)
- EC2 → RDS (fallback, slower)

---

# ⚡ Amazon DynamoDB

---

## 💡 What is it?

- Fully managed NoSQL DB

---

## 🚀 Features

- Serverless
- Auto scaling
- High performance
- Replication across 3 AZs

---

## 📊 Capabilities

- Millions of requests/sec
- Trillions of rows
- Single-digit ms latency

---

## 🔐 Security

- IAM integrated

---

## 💰 Cost

- Low cost
- Pay per usage

---

# ⚡ DynamoDB Accelerator (DAX)

---

## 💡 What is DAX?

- In-memory cache for DynamoDB

---

## 🚀 Benefits

- 10x faster performance
- Microsecond latency

---

## ⚠️ Difference vs ElastiCache

- DAX → Only for DynamoDB
- ElastiCache → Generic caching

---

# 🌍 DynamoDB Global Tables

---

## 💡 What is it?

- Multi-region active-active DB

---

## 🚀 Features

- Read & write in any region
- Low latency globally
- Automatic replication

---

## 🧠 Use Case

- Global applications
- Multi-region failover

---

# 🚨 FINAL REVISION

---

## 🔥 Key Takeaways

- RDS = Managed SQL
- Aurora = High-performance RDS
- Aurora Serverless = Auto-scaling DB
- Read Replica = Scale reads
- Multi-AZ = High availability
- DynamoDB = Serverless NoSQL
- DAX = Cache for DynamoDB
- ElastiCache = General cache
- Global Tables = Multi-region DB

---

# 🚀 Summary

👉 AWS Databases = Fully managed + scalable + highly available
👉 Choose based on:
- SQL → RDS / Aurora
- NoSQL → DynamoDB
- Caching → ElastiCache / DAX
